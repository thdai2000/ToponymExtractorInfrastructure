```python
'''
grouper data structure
{
  "images": {
    "<image_id>": {
      "file_name": "<string>",  // Name of the image file
      "id": "<integer>",       // Unique ID of the image
      "width": "<integer>",    // Image width
      "height": "<integer>"    // Image height
    },
    "...": "..."
  },
  "annotations": {
    "<image_id>": [
      {
        "rec": [list of lists], // List of lists with encoded text
        "bezier": [list of lists] // List of lists with Bezier curve points
      },
      "...": "..."
    ],
    "...": "..."
  },
  "words": {
  "<image_id>": [
    {
        "rec": [list of lists], // List of lists with encoded text
        "bezier": [list of lists], // List of lists with Bezier curve points
        "center": [x,y],
        "group_id": "<integer>",
        "id_in_group": "<integer>",
        "font_embed": [list of floats], // optional
  ]}
}
'''

# merge all grouper dataset
import json
import glob

# Paths to your JSON files
json_files = glob.glob("grouper_data/*.json")

merged_data = {"images": {}, "annotations": {}, "words": {}}
image_id_offset = 0

for json_file in json_files:
    with open(json_file, 'r') as f:
        data = json.load(f)
    
    # Get the current number of images already in merged_data
    current_image_count = len(merged_data["images"])

    # Iterate over images and annotations
    for img_id, image_data in data["images"].items():
        new_img_id = current_image_count + int(img_id)  # Create new image id to avoid collision
        merged_data["images"][str(new_img_id)] = image_data
        merged_data["images"][str(new_img_id)]["id"] = new_img_id  # Update id
        
        # Add annotations and update their reference to the new image ID
        if img_id in data["annotations"]:
            merged_data["annotations"][str(new_img_id)] = data["annotations"][img_id]
        if img_id in data["words"]:
            merged_data["words"][str(new_img_id)] = data["words"][img_id]    

# Save the merged data to a new file
with open("./grouper_data/grouper_data_merged.json", 'w') as out_file:
    json.dump(merged_data, out_file)

print("JSON files merged successfully!")
```
