## simplify annotation polygons

```python
def is_clockwise(poly):
    # Calculate the signed area using the shoelace formula
    x, y = poly.exterior.coords.xy
    area = 0.5 * sum(x[i] * y[i + 1] - x[i + 1] * y[i] for i in range(-1, len(x) - 1))
    return area < 0

def plot_polygons(original: list(), simplified: list()):
    fig, ax = plt.subplots(dpi=300)
    
    # Draw original
    for i, orig in enumerate(original):
        x, y = orig.exterior.xy
        if i == 0:
            ax.plot(x, y, 'b-', label='Original Polygon', linewidth=1.3, alpha=0.7)
        else:
            ax.plot(x, y, 'b-', linewidth=1.3, alpha=0.7)

    # Draw bezier polylines
    for i, beziers in enumerate(simplified):
        
        for j, bezier in enumerate(beziers):
            x, y = bezier.xy
            if (i == 0) and (j == 0):
                ax.plot(x, y, 'r-', label='Bezier curves', linewidth=1., alpha=0.6)
            else:
                ax.plot(x, y, 'r-', linewidth=1., alpha=0.6)

    ax.set_title('Polygon Simplification')
    ax.legend()
    ax.set_aspect('equal')
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.grid(True)
    plt.show()
    
def angle_between(v1, v2):
    '''Computes the angle at a certain point on the polygon based on the two adjacent vertices'''
    v1_u = v1 / np.linalg.norm(v1)
    v2_u = v2 / np.linalg.norm(v2)
    return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))

def get_bezier_polylines(poly):
    if not poly.is_valid or poly.is_empty:
        return None
    
    coords = list(poly.exterior.coords)
    coords = [coords[-2]] + coords
    n = len(coords)
    angles = []
    
    for i in range(1, n-1):
        p0 = np.array(coords[i - 1])  # Last point
        p1 = np.array(coords[i])      # This point
        p2 = np.array(coords[(i + 1) % n])  # Next point
        
        v1 = p1 - p0  # Vect from the last point to the present one
        v2 = p2 - p1  # Vect from the present point to the next one
        
        angle = angle_between(v1, v2)
        angles.append(np.degrees(angle))

    main_angles = sorted(np.argsort(angles)[-4:] + 1)
    
    l1 = LineString(coords[main_angles[0]:main_angles[1]+1])
    l2 = LineString(coords[main_angles[1]:main_angles[2]+1])
    l3 = LineString(coords[main_angles[2]:main_angles[3]+1])
    l4 = LineString(coords[main_angles[3]:-1] + coords[1:main_angles[0]+1])

    segments = np.array([l1, l2, l3, l4], dtype=object)
    longest_segments_arg = np.argsort([l1.length, l2.length, l3.length, l4.length])[-2:]
    beziers = segments[longest_segments_arg]

    return beziers
```

```python
polygons = list()
for i, vertice in enumerate(annotation_polygons):
    poly = Polygon(vertice.copy() + [vertice[0]]).buffer(0).simplify(3)
    if not is_clockwise(poly):
        poly = Polygon(np.flip(list(poly.exterior.coords), 0))
    polygons.append(poly)
    
beziers = []
for polygon in polygons:
    beziers.append(get_bezier_polylines(polygon))
    
# plot_polygons(polygons, beziers)
image = Image.open(img_path)

draw = ImageDraw.Draw(image)

for lines in beziers:
    draw.line([(x, y) for x, y in lines[0].coords] + [(x, y) for x, y in lines[1].coords], fill='blue', width=2)
    
image.show()
```

```python
def get_features(vertices):
    ori_vertices = [(point[0], point[1])for point in vertices]
    poly = Polygon(ori_vertices.copy() + [ori_vertices[0]]).buffer(0).simplify(3)
    if not is_clockwise(poly):
        poly = Polygon(np.flip(list(poly.exterior.coords), 0))
    lines = get_bezier_polylines(poly)
    new_vertices = [(x, y) for x, y in lines[0].coords] + [(x, y) for x, y in lines[1].coords]
    
    feature = (lines[1].coords[0][0] - lines[0].coords[0][0], # x of first point of upper to the first point of lower
               lines[1].coords[0][1] - lines[0].coords[0][1], # y of first point of upper to the first point of lower
               lines[1].coords[-1][0] - lines[0].coords[-1][0], # y of first point of upper to the first point of lower
               lines[1].coords[-1][1] - lines[0].coords[-1][1], # y of first point of upper to the first point of lower
               )
    
    return new_vertices, feature

def find_file_recursive(directory, filename):
    for root, dirs, files in os.walk(directory):
        if filename in files:
            return os.path.join(root, filename)
    return None
```

## matching to correct case

```python
from shapely.geometry import Polygon
from tqdm import tqdm

def intersection_over_union(polygon1_points, polygon2_points):
    """
    Calculate the Intersection over Union (IoU) of two polygons.

    Parameters:
    - polygon1_points: List of (x, y) tuples representing the first polygon.
    - polygon2_points: List of (x, y) tuples representing the second polygon.

    Returns:
    - iou: Intersection over Union metric.
    """
    poly1 = Polygon(polygon1_points)
    poly2 = Polygon(polygon2_points)
    
    if not poly1.is_valid or not poly2.is_valid:
        return 0

    intersection_area = poly1.intersection(poly2).area
    union_area = poly1.union(poly2).area
    
    iou = intersection_area / union_area if union_area != 0 else 0
    
    return iou


def match_polygons(polygon_set1, polygon_set2, iou_threshold=0.5):
    """
    Match polygons from two sets based on IoU.

    Parameters:
    - polygon_set1: List of polygons (each polygon is a list of (x, y) tuples) representing the first set.
    - polygon_set2: List of polygons (each polygon is a list of (x, y) tuples) representing the second set.
    - iou_threshold: Threshold for considering two polygons as a match.

    Returns:
    - matches: List of tuples, each containing the indices of matched polygons (index in set1, index in set2).
    """
    matches = []
    used_indices = set()

    for i, poly1 in tqdm(enumerate(polygon_set1)):
        best_iou = 0  # best intersection over union
        best_match_index = -1
        for j, poly2 in enumerate(polygon_set2):
            if j in used_indices:
                continue
            iou = intersection_over_union(poly1, poly2)
            if iou > best_iou:
                best_iou = iou
                best_match_index = j
        
        if best_iou >= iou_threshold and best_match_index != -1:
            matches.append((i, best_match_index))
            used_indices.add(best_match_index)
    
    return matches

iou_threshold = 0.6
matches = match_polygons(annotation_polygons, model_polygons, iou_threshold)

print(f"Number of Matched polygons: {len(matches)}")
print(f"Number of GT polygons: {len(annotation_polygons)}")
```

```python
# correct the case of text
def uppercase(word):
    if len(word) > 1:
        return word.upper()
    else:
        return word

df_merged['text'] = df_merged['text'].apply(uppercase)
```

```python
# correct the case of text
def changecase(word):
    ocr_text = str(word["OCR_text"])
    annotated_text = str(word["text"])
    returned_str = annotated_text
            
    for i, char in enumerate(ocr_text):
        if char.isupper():
            returned_str = annotated_text[:i] + annotated_text[i:].capitalize()
            try:
                if ocr_text[i+1].isupper():
                    returned_str = annotated_text.upper()
                    break
            except IndexError:
                break
    
    return returned_str

df_merged['text'] = df_merged.apply(changecase, axis=1)

df_merged
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.linear_model import RANSACRegressor
from scipy.interpolate import UnivariateSpline

# Given points
points = np.array([(704.45, 515.5),
 (846.81, 434.15),
 (966.99, 386.08),
 (1133.38, 308.43),
 (1266.5, 265.91),
 (1410.71, 232.63),
 (1553.07, 204.9),
 (1575.26, 221.53),
 (1575.26, 251.12),
 (1416.26, 282.55),
 (1277.59, 317.67),
 (1150.02, 363.9),
 (992.87, 436.0),
 (859.76, 491.47),
 (728.49, 567.27)])

# Apply PCA to find the main axis of the points
pca = PCA(n_components=2)
pca.fit(points)
points_transformed = pca.transform(points)

# Sort points along the new main axis
sorted_indices = np.argsort(points_transformed[:, 0])
points_transformed = points_transformed[sorted_indices]

# Separate the top and bottom points based on the median of the secondary axis
median_y = np.median(points_transformed[:, 1])
top_points = points_transformed[points_transformed[:, 1] <= median_y]
bottom_points = points_transformed[points_transformed[:, 1] > median_y]

def filter_points_ransac(points):
    X = points[:, 0].reshape(-1, 1)
    y = points[:, 1]
    
    ransac = RANSACRegressor(residual_threshold=1)
    ransac.fit(X, y)
    
    inlier_mask = ransac.inlier_mask_
    filtered_points = points[inlier_mask]
    return filtered_points

# Filter points using RANSAC
top_points_filtered = filter_points_ransac(top_points)
bottom_points_filtered = filter_points_ransac(bottom_points)

# Fit splines to the top and bottom points
spline_top = UnivariateSpline(top_points_filtered[:, 0], top_points_filtered[:, 1], k=2, s=1)
spline_bottom = UnivariateSpline(bottom_points_filtered[:, 0], bottom_points_filtered[:, 1], k=2, s=1)

# Generate dense points along the curves
x_dense = np.linspace(points_transformed[:, 0].min(), points_transformed[:, 0].max(), 500)
y_top_dense = spline_top(x_dense)
y_bottom_dense = spline_bottom(x_dense)

# Combine the points into a single array
top_dense_points = np.vstack((x_dense, y_top_dense)).T
bottom_dense_points = np.vstack((x_dense, y_bottom_dense)).T[::-1]  # Reverse to maintain clockwise order
polygon_points_transformed = np.vstack((top_dense_points, bottom_dense_points))

# Transform points back to the original coordinate system
polygon_points = pca.inverse_transform(polygon_points_transformed)
top_points_filtered = pca.inverse_transform(top_points_filtered)
bottom_points_filtered = pca.inverse_transform(bottom_points_filtered)
top_points = pca.inverse_transform(top_points)
bottom_points = pca.inverse_transform(bottom_points)

# Plot the results
plt.figure(figsize=(10, 8))
plt.scatter(top_points[:, 0], top_points[:, 1], color='grey', label='Original Points')
plt.scatter(bottom_points[:, 0], bottom_points[:, 1], color='grey', label='Original Points')
plt.scatter(top_points_filtered[:, 0], top_points_filtered[:, 1], color='red', label='Original Points')
plt.scatter(bottom_points_filtered[:, 0], bottom_points_filtered[:, 1], color='orange', label='Original Points')
plt.plot(polygon_points[:, 0], polygon_points[:, 1], color='blue', label='Estimated Polygon')
plt.legend()
plt.title('Estimated Polygon Wrapping Text with Geometric Feature Filtering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
